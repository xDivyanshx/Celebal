```python
import matplotlib.pyplot as plt
import pandas as pd
#not using seaborn as was not part of the given course
```

```python
data = pd.read_csv("Online Sales Data.csv")
print(data.head())
```

```python
#Line Plot to show Total Revenue over Time
data['Date'] = pd.to_datetime(data['Date'])
revenue_over_time = data.groupby('Date')['Total Revenue'].sum()
plt.figure(figsize=(12, 6))
plt.plot(revenue_over_time.index, revenue_over_time.values, marker='o')
plt.title('Total Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.show()
```

```python
#Bar Chart to show Total Units sold per Category
units_sold_per_category = data.groupby('Product Category')['Units Sold'].sum()
plt.figure(figsize=(12, 6))
units_sold_per_category.plot(kind='bar', color='skyblue')
plt.title('Total Units Sold per Product Category')
plt.xlabel('Product Category')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
#Pie Chart to show distribution of Sales by Region
sales_by_region = data.groupby('Region')['Total Revenue'].sum()
plt.figure(figsize=(8, 8))
sales_by_region.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Sales by Region')
plt.ylabel('')
plt.show()
```

```python
#Pie Chart to show Distribution of Units Sold by Region
region_units_sold = data.groupby('Region')['Units Sold'].sum()
plt.figure(figsize=(10, 6))
plt.pie(region_units_sold, labels=region_units_sold.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribution of Units Sold by Region')
plt.axis('equal')
plt.show()
```

```python
#Pie Chart to show distribution of Sales by Product Category
revenue_by_category = data.groupby('Product Category')['Total Revenue'].sum()
plt.figure(figsize=(10, 7))
plt.pie(revenue_by_category, labels=revenue_by_category.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(revenue_by_category))))
plt.title('Distribution of Sales by Product Category')
plt.axis('equal')
plt.show()
```

```python
#Bar Chart to show distribution of Payment Methods
payment_method_counts = data['Payment Method'].value_counts()
plt.figure(figsize=(12, 6))
payment_method_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Payment Methods')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Scatter Plot show relatiob between Unit Price and its Units sold
plt.figure(figsize=(12, 6))
plt.scatter(data['Unit Price'], data['Units Sold'], alpha=0.5, edgecolors='w', s=100)
plt.title('Unit Price vs. Units Sold')
plt.xlabel('Unit Price')
plt.ylabel('Units Sold')
plt.grid(True)
plt.show()
```

```python
#Bar chart to show the five most sold products
units_sold_by_product = data.groupby('Product Name')['Units Sold'].sum()
top_five_products = units_sold_by_product.sort_values(ascending=False).head(5)
plt.figure(figsize=(12, 6))
top_five_products.plot(kind='bar', color='skyblue')
plt.title('Top 5 Most Sold Products')
plt.xlabel('Product Name')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
#Grouped Bar Chart to show different payment methods used according to region
payment_method_by_region = data.groupby(['Region', 'Payment Method']).size().unstack()
payment_method_by_region.plot(kind='bar', figsize=(14, 7))
plt.title('Distribution of Payment Methods by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Payment Method', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#Stacked Bar Chart to show Product Categories based on Region
category_by_region = data.groupby(['Region', 'Product Category']).size().unstack()
category_by_region.plot(kind='barh', stacked=True, figsize=(12, 8))
plt.title('Distribution of Product Categories by Region')
plt.xlabel('Count')
plt.ylabel('Region')
plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
